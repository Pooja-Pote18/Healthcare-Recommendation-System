```python
import pandas as pd
import numpy as np
import joblib

data = pd.read_excel(r"C:\Users\Simran Gurav\OneDrive\Desktop\data science Project\Healthcare_Project\Healthcare System\dataset\healthcare_recommendation_dataset.xlsx")

print(data.head())
```

```python
print(data.columns)
```

```python
#Feature Selection

X = data[['age','blood_pressure','glucose','heart_rate']]
y = data['diagnosis']
```

```python
#Train-Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)

print(predictions)
```

```python
#Model Accuracy 

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)

print("Model Accuracy:", accuracy)
```
